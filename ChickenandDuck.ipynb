Code to Download Images

```python
import os
from icrawler.builtin import GoogleImageCrawler

def download_images(keyword, folder, max_num=1000):
    os.makedirs(folder, exist_ok=True)
    crawler = GoogleImageCrawler(storage={'root_dir': folder})
    crawler.crawl(keyword=keyword, max_num=max_num)
    print(f"✅ Downloaded {max_num} images for '{keyword}'")

# Download Duck and Chicken images
download_images("duck bird", "dataset/duck", max_num=1000)
download_images("wild duck bird", "dataset/duck", max_num=1000)
download_images("white duck", "dataset/duck", max_num=1000)
download_images("chicken bird", "dataset/chicken", max_num=1000)
download_images("hen bird", "dataset/chicken", max_num=1000)
download_images("chicken farm", "dataset/chicken", max_num=1000)


```

Imports and Setup

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from sklearn.metrics import classification_report
import splitfolders
import matplotlib.pyplot as plt
import numpy as np
import os
```

Dataset Preparation

```python
import splitfolders

# Split the data
splitfolders.ratio('dataset',   # This is your original folder with 'duck' and 'chicken' inside
                   output="dataset_split",
                   seed=1337,
                   ratio=(0.7, 0.2, 0.1))
# 70% train, 20% validation, 10% test
```

```python
# Assuming you already have:
# dataset_split/
#   ├── train/
#   ├── val/
#   └── test/

train_dir = '/content/dataset_split/train'
val_dir = '/content/dataset_split/val'
test_dir = '/content/dataset_split/test'

IMG_SIZE = 224
BATCH_SIZE = 32
```

Data Generators

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(IMG_SIZE, IMG_SIZE),
                                                    batch_size=BATCH_SIZE,
                                                    class_mode='binary')

val_generator = val_datagen.flow_from_directory(val_dir,
                                                target_size=(IMG_SIZE, IMG_SIZE),
                                                batch_size=BATCH_SIZE,
                                                class_mode='binary')

test_generator = val_datagen.flow_from_directory(test_dir,
                                                 target_size=(IMG_SIZE, IMG_SIZE),
                                                 batch_size=1,
                                                 class_mode='binary',
                                                 shuffle=False)
```

Load MobileNetV2 (pre-trained), Add Head

```python
# Load pre-trained model without top layer
base_model = tf.keras.applications.MobileNetV2(input_shape=(IMG_SIZE, IMG_SIZE, 3),
                                               include_top=False,
                                               weights='imagenet')

base_model.trainable = False  # Freeze base initially

# Add custom head
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(64, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

Train New Head Only

```python
initial_epochs = 5

history = model.fit(train_generator,
                    validation_data=val_generator,
                    epochs=initial_epochs)
```

Unfreeze Some Base Layers for Fine-Tuning

```python
# Unfreeze the base model
base_model.trainable = True

# Optionally: Freeze first few layers, unfreeze last 20
for layer in base_model.layers[:-20]:
    layer.trainable = False

# Recompile with low learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Fine-Tune Whole Model

```python
fine_tune_epochs = 5
total_epochs = initial_epochs + fine_tune_epochs

history_fine = model.fit(train_generator,
                         validation_data=val_generator,
                         epochs=total_epochs,
                         initial_epoch=history.epoch[-1])
```

Final Evaluation (Classification Report)


```python
# Predict
y_pred_probs = model.predict(test_generator)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()
y_true = test_generator.classes

# Report
print(classification_report(y_true, y_pred, target_names=['chicken', 'duck']))
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Get predictions (you probably already have y_pred and y_true)
y_pred_probs = model.predict(test_generator)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()
y_true = test_generator.classes

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Class names
class_names = ['chicken', 'duck']

# Plot
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Duck vs Chicken Classification')
plt.show()
```
